```python
dfc = pd.read_csv("Data file.csv",sep=';')
missing_values = dfc.isnull().sum()


if missing_values.any():
    for column, count in missing_values.items():
        if count > 0:
            print(f"{column}: {count}")
else:
    print("No missing values detected")
```

```python
outliers = {}


for column in dfc.columns:
      
        Q1 = dfc[column].quantile(0.25)
        Q3 = dfc[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_count = ((dfc[column] < lower_bound) | (dfc[column] > upper_bound)).sum()
        outliers[column] = outlier_count
print(outliers)
```

```python

df = pd.read_csv("Data file.csv",sep=';')

for column in df.columns:
    min_v=df[column].min()
    max_v=df[column].max()
    df[column]=df[column].transform(lambda x: (x-min_v) / (max_v-min_v))
df
       
        
        
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import metrics


for column in df.columns:
    x=df[column].mean()
    df[column].fillna(x,inplace=True)




n, m = df.shape
target = df[df.columns[-1]]
x = df.loc[:, :df.columns[-2]]

lm = LinearRegression()
model = lm.fit(x, target)

ypred = lm.predict(x)


mse = metrics.mean_squared_error(target, ypred)
r2 = metrics.r2_score(target, ypred)

# Sonuçları yazdırma
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.scatter(target, ypred, edgecolors='k', alpha=0.7)
plt.plot([target.min(), target.max()], [target.min(), target.max()], 'k--', lw=2)
plt.xlabel('Actual Wine Quality Ratings')
plt.ylabel('Predicted Wine Quality Ratings')
plt.title('Actual vs Predicted Wine Quality Ratings')
plt.grid(True)
plt.show()
```
